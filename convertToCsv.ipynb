```python
import pandas as pd

# Read the original CSV file
df = pd.read_csv('grouped_dataset.csv')

# Initialize an empty dictionary to keep track of items
item_dict = {}

# Iterate through each row in the dataframe
for index, row in df.iterrows():
    items_list = eval(row['Item Name'].replace("'", '"'))  # Replace single quotes with double quotes and convert the string representation to a list
    
    # Update the item_dict for each item in the list
    for item in items_list:
        if item not in item_dict:
            item_dict[item] = [''] * len(df)
        item_dict[item][index] = 't'

# Create a new dataframe using the item_dict
result_df = pd.DataFrame(item_dict)

# Save the result to a new CSV file without 'Item Name' column
result_df.to_csv('result_file.csv', index=False)
```
